```python
from pyspark.sql import SparkSession
```

```python
spark=SparkSession.builder.appName('k-means_first_example').getOrCreateOrCreate()
```

```python
input_file_path="file:///C:/Users/ckp43_000/Documents/a2a.t"
```

```python
data=spark.read.format('libsvm').load(input_file_path)
```

```python
data.show(5)
```

```python
data.count()
```

```python
final_data=data.select('features')
```

```python
final_data.show(5)
```

```python
from pyspark.ml.clustering import KMeans,KMeansModel
```

```python
kmeans=KMeans().setK(4).setSeed(10)
```

```python
model=kmeans.fit(final_data)
```

```python
centres=model.clusterCenters()
```

```python
centres
```

```python
result=model.transform(final_data)
```
